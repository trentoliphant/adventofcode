```python
room = {}
x = 0
y = 0
dirs = {
    1: (0, 1),
    2: (0, -1),
    3: (-1, 0),
    4: (1, 0)
}
```

```python
def move(dir):
    global x
    global y
    robot.add_input(dir)
    robot.run()
    status = robot.output[-1]
    xmove = dirs[dir][0]
    ymove = dirs[dir][1]
    if status == 0:
        room[(x + xmove, y + ymove)] = status
    else:
        x += xmove
        y += ymove
        room[(x + xmove, y + ymove)] = status
    return x, y, status
```

```python
for i in range(1000):
    x, y, object = move(1)
    if object == 4:
        print("Found a wall", x, y)
        break
```
