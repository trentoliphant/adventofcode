```python
def get_blockers(x, y, x1, y1):
    xdir = 1 
    ydir = 1
    if x1 < x:
        xdir = -1
    if y1 < y:
        ydir = -1
    x_diff = abs(x - x1)
    y_diff = abs(y - y1)
    blocked_by = []
    #diagonal
    if x_diff == y_diff and x_diff > 1:
        for i in range(1, x_diff):
            blocked_by.append((x + i * xdir, y + i * ydir))
    #horizontal
    elif x_diff == 0 and y_diff > 1:
        for i in range(1, y_diff):
            blocked_by.append((x, y + i * ydir))
    #vertical
    elif y_diff == 0 and x_diff > 1:
        for i in range(1, x_diff):
            blocked_by.append((x + i * xdir, y))
    #others
    elif y_diff > 1 and x_diff > 1:
        if x_diff > y_diff:
            max_diff = x_diff
        else:
            max_diff = y_diff
#        print(F'x_diff = {x_diff}, y_diff = {y_diff}, max_diff = {max_diff}')
        for i in range(2, max_diff):
            #find Least Common Divisor
            if x_diff % i  == 0 and y_diff % i == 0:
                x_base = x_diff//i
                y_base = y_diff//i
                for j in range(2, min(x_base, y_base) + 1):
#                    print(j, x_base % j, y_base % j)
                    if x_base % j == 0 and y_base % j == 0:
                        x_base = x_base//j
                        y_base = y_base//j
#                print(F'x_base = {x_base}, y_base = {y_base}')
#                print(F'max_diff: /x_base = {max_diff/x_base}, /y_base = {max_diff/y_base}')
                for j in range(1, max_diff//max(x_base, y_base)):
#                    print(j)
                    blocked_by.append((x + x_base * j * xdir, y + y_base * j * ydir))
                break
    return blocked_by
```

```python
def get_angle(x, y, x1, y1, flipx=False, flipy=False):
    x_diff = x1 - x
    y_diff = y1 - y
    if flipx:
        x_diff = x_diff * -1
    if flipy:
        y_diff = y_diff * -1
    rad = math.atan2(x_diff, y_diff)
    if rad < 0:
        rad += 2 * math.pi
    return rad
```

```python
xrange = 36
yrange = 36
x = 25
y = 31

angles = {}
blocked_count = {}
can_see = []
for x1 in range(xrange):
    for y1 in range(yrange):
        if (x1,y1) == (x,y):
            angle = -1
            pass
        elif (x1,y1) in asteroids:
            blockers = set(get_blockers(x,y,x1,y1)) & set(asteroids)
            angle = get_angle(x, y, x1, y1, flipy=True)
            if blockers:
                blocked_count[(x1, y1)] = len(blockers)
                angle += len(blockers) * 2 * math.pi
            else:
                can_see.append((x1, y1))
            angles[(x1, y1)] = angle
#            print(x1, y1, x1-x,y1-y, angle, blockers)
```
