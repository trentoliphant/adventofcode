```python
input_data = open('data/day07.txt').read().split('\n')
```

```python
def parse_rule(rule):
    rule = rule.replace(' bags','').replace(' bag','').replace('.','')
    (out_bag, in_bags) = rule.split(' contain ')
    if in_bags == 'no other':
        return (out_bag, [])
    in_bags = in_bags.split(', ')
    return (out_bag, [(bag[2:], int(bag[0])) for bag in in_bags])
```

```python
inners = {}
outers = {}

for rule in input_data:
    outer, inner = parse_rule(rule)
    outers[outer] = inner
    for i in inner:
        if inners.get(i[0]):
            inners[i[0]].append(outer)
        else:
            inners[i[0]] = [outer]
```

```python
def get_inner_count(outer, initial=False):
    inner_bags = outers[outer]
    if len(inner_bags) == 0:
        return 1
    if initial:
        bag_count = 0
    else:
        bag_count = 1
    for (inner, incount) in inner_bags:
        bag_count += incount * get_inner_count(inner)
    return bag_count
        
```

```python
get_inner_count('shiny gold', True)
```
