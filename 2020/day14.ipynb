```python
def apply_mask(value, mask):
    val_bstr = format(value, '036b')
    new_str = ''
    for i in range(36):
        if mask[i] != 'X':
            new_let = mask[i]
        else:
            new_let = val_bstr[i]
        new_str += new_let
    return int(new_str, 2)
            
```

```python
def get_masked_address(addr, mask):
    addr_bstr = format(addr, '036b')
    new_str = ''
    for i in range(36):
        if mask[i] == '0':
            new_str += addr_bstr[i]
        else:
            new_str += mask[i]
        
    return new_str
```

```python
def get_addresses(mask):
    addresses = set()
    mask_bases = ['']
    for idx, letter in enumerate(mask):
        for i in range(len(mask_bases)):
            if letter != 'X':
                mask_bases[i] = mask_bases[i] + letter
            else:
                base = mask_bases[i]
                new_base_0 = base + "0"
                new_base_1 = base + "1"
                mask_bases[i] = new_base_0
                mask_bases.append(new_base_1)
                rem = mask[idx+1:]
                rem_0 = rem.replace('X', '0')
                rem_1 = rem.replace('X', '1')
                addresses.add(base + '0' + rem_0)
                addresses.add(base + '1' + rem_0)
                if rem_0 != rem_1:
                    addresses.add(base + '0' + rem_1)
                    addresses.add(base + '1' + rem_0)
    return addresses
```

```python
# Part 2
mem = {}
cur_mask = '0'*36
for inst in init:
    act, value = inst.split(' = ')
    if act == 'mask':
        cur_mask = value
    else:
        addr = act[4:-1]
        new_mask = get_masked_address(int(addr), cur_mask)
        for a in get_addresses(new_mask):
            mem[a] = int(value)
```
