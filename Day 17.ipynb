```python
rows = map.split('\n')
```

```python
grid = []
for row in rows:
    grid.append([col for col in row])
```

```python
def find_intersections(grid):
    intersections = []
    for x, row in enumerate(grid[1:-3]):
        #print(x, ''.join(row[1:-1]))
        for y, col in enumerate(row[1:-1]):
            if col == '#':
                intersects = []
         #       print(x+1,y+1, grid[x+1][y+1])
                for dir in [(0,1), (1,0), (0,-1), (-1, 0)]:
                    x1 = x + 1 + dir[0]
                    y1 = y + 1 + dir[1]
           #         print('\t',x1, y1, grid[x1][y1])
                    intersects.append((grid[x1][y1] == '#'))
          #      print(intersects)
                if all(intersects):
                    intersections.append((x+1, y+1))
    return intersections
```

```python
align = find_intersections(grid)
```

```python
align
```

```python
sum = 0
for point in align:
    sum += (point[0]*point[1])
```
