```python
adjacents = {
    0: [1,5],
    1: [0,2,6],
    2: [1,3,7],
    3: [2,4,8],
    4: [3,9],
    5: [0,6,10],
    6: [1,5,7,11],
    7: [2,6,8,12],
    8: [3,7,9,13],
    9: [4,8,14],
    10: [5,11,15],
    11: [6,10,12,16],
    12: [7,11,13,17],
    13: [8,12,14,18],
    14: [9,13,19],
    15: [10,16,20],
    16: [11,15,17,21],
    17: [12,16,18,22],
    18: [13,17,19,23],
    19: [14,18,24],
    20: [15,21],
    21: [16,20,22],
    22: [17,21,23],
    23: [18,22,24],
    24: [19,23]
}

map = '''
#####
...##
#..#.
#....
#...#
'''.replace('\n','')
```

```python
bugs = [c == '#' for c in map]
```

```python
bugs
```

```python
def has_bug(space, map, verbose=False):
    if verbose:
        print(F'Current status: {map[space]}')
    count = 0
    for x in adjacents[space]:
        if verbose:
            print("Checking", x)
        if map[x]:
            if verbose:
                print('found bug')
            count += 1
    if verbose:
        print(F'Total adjacent: {count}')
    if map[space] and count != 1:
        if verbose:
            print('Bug dies')
        return False
    elif not map[space] and (count == 1 or count==2):
        if verbose:
            print('Bug spawns')
        return True
    else:
        if verbose:
            print('No change')
        return map[space]
```

```python
def update_map(map):
    new_map = [has_bug(i, map) for i in range(len(map))]
    return new_map
```

```python
def print_map(map):
    for x in range(5):
        row = ['#' if x else '.' for x in map[x*5:x*5+5]]
        print(''.join(row))
```

```python
has_bug(0,bugs,True)
```

```python
seen = []
```

```python
cur_state = bugs[:]
while cur_state not in seen:
    seen.append(cur_state)
    cur_state = update_map(cur_state)
    print_map(cur_state)
    print()
```
